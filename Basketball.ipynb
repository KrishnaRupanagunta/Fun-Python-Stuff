# Library import

```python
#import python libraries 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Generate sequences

```python
std_dev=[]
sigma=0;
mu=0;
std_mean=[]
p_x=[]
p=0.60;

while p<1.0:
  mytrials=[];
  trial_ct=0;
  seq=0;
  ct=0;
  while trial_ct<1000:
    ct=0;
    seq=0;
    while seq<5:
      sf=np.random.binomial(1,p);
      if sf==1:
        seq+=1
      else:
        seq=0
      ct+=1
    mytrials.append(ct);
    trial_ct+=1;
    sigma=round(np.std(mytrials),2);
    mu=round(np.mean(mytrials),2);
  std_dev.append(sigma);
  std_mean.append(mu);
  p_x.append(p);
  if p==0.70: 
   plt.hist(mytrials,bins='auto',label=True);
   plt.ylabel('Count');
   plt.xlabel('#of Attempts')
   string_stuff = r'For p=0.7: $Mean={}, SD={}$'.format(mu,sigma)
   plt.text(30,60, string_stuff)
   plt.show();
   print();
  p+=0.05;
  p=round(p,2);

x_cord=[0.60,0.95];
y_cord=[5,5];
string_stuff='Change with skill-level (increasing p)'
plt.plot(p_x,std_mean,label="Mean");
plt.plot(p_x,std_dev, label="Std. Dev.");
plt.plot(x_cord,y_cord,color='black',linestyle='dashed');
plt.title('Change with skill-level (increaing p)');
plt.legend();
plt.show();
```
